PROBLEM 1 Taxi Fare

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = {
    'trip_miles': [1.2, 2.0, 3.5, 4.1, 5.6, 6.3, 0.9, 2.5, 3.0, 7.0, 1.8, 4.5, 6.8, 2.2, 5.0],
    'trip_seconds': [240, 360, 650, 780, 1000, 1150, 200, 420, 540, 1300, 300, 900, 1280, 400, 950],
    'pickup_community_area': [8, 5, 12, 15, 9, 22, 11, 7, 16, 3, 10, 20, 6, 13, 2],
    'dropoff_community_area': [10, 7, 14, 18, 12, 27, 13, 9, 20, 5, 15, 25, 4, 8, 6],
    'fare': [6.5, 9.0, 13.5, 16.0, 19.8, 21.5, 5.0, 10.2, 12.0, 24.0, 7.5, 17.8, 23.5, 9.5, 18.0]
}
df = pd.DataFrame(data)      #CH.SC.U4CSE23223
```

```python
X = df[['trip_miles', 'trip_seconds', 'pickup_community_area', 'dropoff_community_area']]
y = df['fare']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)     #CH.SC.U4CSE23223
```

```python
print("Predicted Fares:", y_pred)
print("Actual Fares:   ", y_test.values)
print("R² Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
